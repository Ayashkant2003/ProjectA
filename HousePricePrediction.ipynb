## Observations

### 1. n_hos_beds column has missing values
### 2. Distribution of crime_rate is right skewed
### 3. n_hot_rooms column has 2 outliers
### 4. rainfall column has 1 outlier
### 5. bus_ter column has only YES values

```python
# Filling the missing data

df['n_hos_beds'] = df['n_hos_beds'].fillna(df['n_hos_beds'].mean())
```

```python
df.info()
```

```python
# Reducing the skewness of crime_rate data


df['crime_rate']=np.log(1+df['crime_rate'])
```

```python
sns.scatterplot(x='crime_rate',y='price',data=df)
```

```python
# Treating the outliers of n_hot_rooms

# Inter Quartile Range

Q1 = df['n_hot_rooms'].quantile(0.25)
Q3 = df['n_hot_rooms'].quantile(0.75)

IQR =Q3-Q1

upper_bound = Q3+1.5*IQR

```

```python
df[df['n_hot_rooms']>upper_bound]
```

```python
df.loc[df['n_hot_rooms'] > upper_bound, 'n_hot_rooms'] = upper_bound
```

```python
sns.jointplot(x='n_hot_rooms',y='price',data=df)
```

```python
# Treating outlier of rainfall columns

lv=np.percentile(df.rainfall,1)
lv
```

```python
df.loc[df['rainfall']<0.5*lv,'rainfall']=0.5*lv
```

```python
sns.scatterplot(x='rainfall',y='price',data=df)
```

## Implementing Neural Networks for price prediction

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
X=df.drop('price',axis=1).values
```

```python
y=df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scaling the data 

scaler=MinMaxScaler()

X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
#Creating the model

model=Sequential()


model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='relu'))

model.compile(loss='mse',optimizer='adam')
```

```python
# Fitting the Data

model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
# Evaluationg the model

pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
print('MAE:', mean_absolute_error(y_test, pred))  
print('MSE:', mean_squared_error(y_test, pred))  
print('RMSE:', np.sqrt(mean_squared_error(y_test, pred)))
print('VarScore:', explained_variance_score(y_test, pred))
```
